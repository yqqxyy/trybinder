```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
col1 = np.random.randn(100,1)
col2 = np.random.rand(100,1)
alldata = np.concatenate((col1,col2),axis=1)
testpd = pd.DataFrame(alldata,columns=['data1','data2'],index=np.arange(0,1,0.01))
testpd.plot()
```

```python
testpd['data3'] = (testpd.data1+10)*5
plt.hist(testpd['data3'],10)
```
